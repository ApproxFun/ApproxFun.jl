# $$u'' + 2u = \cos(\omega t)\qquad u(0)=u'(0)=0$$ $\omega=\sqrt 2$ is a resonance and the solution grows

```julia
t=Fun(identity,[0.,1000.]) 
D=Derivative()           
B=[ldirichlet(),lneumann()]  

L=D^2+2I  # our differential operator

@manipulate for ω=0.:.05sqrt(2):2sqrt(2)
    u=[B,L]\[0.,0.,cos(ω*t)]
    ApproxFun.plot(u)
end


```

# Bessel equation $$x^2 u'' + x u + (x^2 - \nu^2)u$$

```julia
x=Fun(identity,[1.,1000.])  
d=domain(x)                    
D=Derivative(d)               
B=dirichlet(d)  
@manipulate for ν=0:.1:1000
	L=x^2*D^2 + x*D + (x^2 - ν^2)   # our differential operator
   	u=[B,L]\[besselj(ν,first(d)),besselj(ν,last(d))]
   	ApproxFun.plot(u;axis=[-.2,.2])
end
```

# $$\Delta u + k u = 0, u(\pm 1,y)=u(x,\pm 1) = 1$$

```julia
d=Interval()^2
B=dirichlet(d)
Δ=lap(d)

@manipulate for k=-500.0:.001:2000.0,ny=10:200
    ApproxFun.contour(pdesolve([B,Δ+k*I],ones(4),ny))
end
```
